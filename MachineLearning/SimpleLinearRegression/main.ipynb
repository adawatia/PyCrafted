```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib  inline
```

```python
df = pd.read_csv('HeightWeight.csv')
```

```python
df.head()
```

```python
#Scatter Plot 
plt.scatter( y = df['Height(Inches)'],x = df['Weight(Pounds)'])
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
# Corelation
df.corr()
```

```python
sns.pairplot(df)
```

```python
## Independent  and Dependent features
X = df[['Weight(Pounds)']] ## independent feature should be a df or 2d array
y = df['Height(Inches)'] ## dependent feature can be in series or 1d array
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
## Standarization 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
## Data Leakage
X_test = scaler.transform(X_test)
```

```python
#  Simple Linear Regrssion
from sklearn.linear_model import LinearRegression
```

```python
regresstion = LinearRegression(n_jobs=-1)
```

```python
regresstion.fit(X_train, y_train)
```

```python
print(f"Coefficient or slope: {regresstion.coef_}")
print(f"Intercept: {regresstion.intercept_}")
```

```python
##  PLot best fit line using training data
plt.scatter(X_train,y_train)
plt.plot(X_train,regresstion.predict(X_train))
```

```python
## Prediction for test data
y_pred = regresstion.predict(X_test)
```

```python
## Performance Metrices
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```
